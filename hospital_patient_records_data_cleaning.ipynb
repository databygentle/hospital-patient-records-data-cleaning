```python
# to import the python library: pandas for this healthcare dataset helps me identify issues , address/solve it and validate/verify;numpy for working with numbers and arrays faster 
import pandas as pd
import numpy as np
```

```python
# to load the data set into the notebook, the dateset contains input variables used to predict the stay time of a patient.
df = pd.read_csv("train_data.csv")
df
```

```python
# this shows the first five row in the data set, to get an overview of the data and understand what information it contains
df.head()
```

```python
# to get the information of the data set like its structure, it contents and the data types of each columns.
df.info()
```

```python
#  to get the statistical summary of the data set and knowing the distribution of each column to determine which imputation method to use
df.describe()
```

```python
# to calculate the percentage of missing values in its respective column
df.isnull().sum() / len(df) * 100
```

```python
# to check if there is any duplicated row
df.duplicated().sum()
```

```python
# Mean imputation for normally distributed column
df['Bed Grade'] = df['Bed Grade'].fillna(df['Bed Grade'].mean())
```

```python
# Median imputation for skewed columns - patientid, City_Code_Patient, and Visitors with Patient has extreme value
df['patientid'] = df['patientid'].fillna(df['patientid'].median())
df['City_Code_Patient'] = df['City_Code_Patient'].fillna(df['City_Code_Patient'].median())
df['Visitors with Patient'] = df['Visitors with Patient'].fillna(df['Visitors with Patient'].median())
```

```python
# Mode imputation for categorical columns
df['Type of Admission'] = df['Type of Admission'].fillna(df['Type of Admission'].mode()[0])
df['Severity of Illness'] = df['Severity of Illness'].fillna(df['Severity of Illness'].mode()[0])
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
df['Stay'] = df['Stay'].fillna(df['Stay'].mode()[0])
```

```python
# to verify if the full data set is clean
df.isnull().sum()
```

```python
# to convert the data type of Bed Grade & City_Code_Patient column from float to integer
df['Bed Grade'] = df['Bed Grade'].astype(int)
df['City_Code_Patient'] = df['City_Code_Patient'].astype(int)
df
```
